```python
#Step 2 - Compute Mean and Covariance Matrxi
mean=np.mean(data,axis=0)
cov=np.cov(data,rowvar=False)
inv_cov=np.linalg.inv(cov)
print(mean)
print(cov)
print(inv_cov)
```

```python
#Step 3 - Compute MD for each point
md=[distance.mahalanobis(x,mean,inv_cov)for x in data]
md=np.array(md)
print(md)
```

```python
#Step 4 - Visualize the MD
plt.figure(figsize=(7,6))
plt.scatter(data[:,0],data[:,1],c=md,cmap='viridis')
plt.colorbar(label='Mahalanobis distance')
plt.xlabel('X1')
plt.ylabel('Y1')
plt.title('Mahalanobis distance for Each Point')
plt.show()
```

```python
# Step 1 - Generate sample 2D data
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance

# Set random seed for reproducibility
np.random.seed(0)

# Sample 2D data points
data = np.array([
    [170, 65],
    [165, 60],
    [180, 75],
    [175, 70]
])

print("Data shape:", data.shape)

# Visualize 2D dataset
plt.figure(figsize=(6, 5))
plt.scatter(data[:, 0], data[:, 1], color='blue', s=100)
plt.xlabel('X1')
plt.ylabel('Y1')
plt.title('2D Dataset (Scattered Dots)')
plt.grid(True)
plt.show()

# Step 2 - Compute Mean and Covariance Matrix
mean = np.mean(data, axis=0)
cov = np.cov(data, rowvar=False)
cov = cov + np.eye(cov.shape[0]) * 1e-6  # Regularization
inv_cov = np.linalg.inv(cov)

print("\nMean:\n", mean)
print("\nCovariance Matrix:\n", cov)
print("\nInverse Covariance Matrix:\n", inv_cov)

# Step 3 - Compute Mahalanobis Distance for each point
md = [distance.mahalanobis(x, mean, inv_cov) for x in data]
md = np.array(md)
print("\nMahalanobis Distances:\n", md)

# Step 4 - Visualize Mahalanobis Distance (as colored scattered dots)
plt.figure(figsize=(7, 6))
scatter = plt.scatter(data[:, 0], data[:, 1],
                      c=md, cmap='viridis', s=180, edgecolors='black')
plt.colorbar(scatter, label='Mahalanobis Distance')
plt.xlabel('X1')
plt.ylabel('Y1')
plt.title('Mahalanobis Distance Visualization (Scattered Dots)')
plt.grid(True)
plt.show()
```
